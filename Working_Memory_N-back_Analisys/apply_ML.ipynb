```python
import pandas as pd #  manipulação e análise de dados. Ooferece estruturas de dados poderosas, como DataFrames.
import matplotlib.pyplot as plt # usada para visualização de dados em Python. É útil para análise exploratória de dados.

from sklearn.preprocessing import StandardScaler # Técnica de pré-processamento comum para padronizar as features removendo a média e escalando para a variância unitária.

from sklearn.decomposition import PCA # Técnica de redução de dimensionalidade que ajuda a encontrar as componentes principais em um conjunto de dados.

from sklearn.neighbors import KNeighborsClassifier # Algoritmo de aprendizado supervisionado para classificação baseado na proximidade dos exemplos no espaço de features.
from sklearn.cluster import KMeans #  algoritmo de clustering não supervisionado que agrupa os dados em clusters baseados na similaridade das features.

from sklearn.model_selection import train_test_split # É uma função para dividir o conjunto de dados em conjuntos de treinamento e teste.
from sklearn.model_selection import GridSearchCV # É uma técnica de otimização de hiperparâmetros que realiza uma busca exaustiva pelos melhores hiperparâmetros para um estimador

# São métricas de avaliação usadas para medir o desempenho de modelos de machine learning:
from sklearn.metrics import accuracy_score, silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score, confusion_matrix

from google.colab import drive
drive.mount('/content/drive/')
%cd /content/drive/MyDrive/
```

```python
caminho_arquivo = '/content/drive/MyDrive/Atividade Final - Memória de trabalho EEG/Analise_Dados/Registros_Openvibe_recoleta/signal_ml_.csv'
raw_data = pd.read_csv(caminho_arquivo)

# Remover a coluna 'n_janela'
raw_data.drop('n_janela', axis=1, inplace=True)

# Renomear a coluna 'n_back' para 'target'
raw_data.rename(columns={'n_back': 'target'}, inplace=True)

# Substituir os valores NaN na coluna 'target' por 0
raw_data['target'].fillna(0, inplace=True)

raw_data
```

```python
print("grupo 1:", len(raw_data[raw_data["target"]==1]))
print("grupo 2:", len(raw_data[raw_data["target"]==2]))
print("grupo 3:", len(raw_data[raw_data["target"]==3]))
print("grupo 0:", len(raw_data[raw_data["target"]==0]))
```

```python
grupo_1 = raw_data[raw_data["target"]==1].copy()
grupo_2 = raw_data[raw_data["target"]==2].copy()
grupo_3 = raw_data[raw_data["target"]==3].copy()
grupo_0 = raw_data[raw_data["target"]==0].copy()
```

```python
from sklearn.utils import resample

# Balanceamento do grupo 1
grupo_1_balanced = resample(grupo_1, replace=True, n_samples=34816, random_state=42)

# Balanceamento do grupo 2
grupo_2_balanced = resample(grupo_2, replace=True, n_samples=34816, random_state=42)

# Balanceamento do grupo 3
grupo_3_balanced = resample(grupo_3, replace=True, n_samples=34816, random_state=42)

# Combinando os grupos balanceados
dados_balanceados = pd.concat([grupo_0, grupo_1_balanced, grupo_2_balanced, grupo_3_balanced])
```

```python
# Normalizar os dados antes de aplicar o PCA
scaler = StandardScaler()
dados_normalizados = scaler.fit_transform(dados_balanceados)

# Aplicar o PCA para reduzir a dimensionalidade
pca = PCA(n_components=4)  # Defina o número de componentes desejado
dados_reduzidos = pca.fit_transform(dados_normalizados)
```

```python
# Dividir os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(dados_reduzidos, dados_balanceados['target'], test_size=0.2, random_state=42)

# Treinar um classificador KNeighbors
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
# Fazer previsões no conjunto de teste
predicoes = knn.predict(X_test)

# Calcular a acurácia do modelo
acuracia = accuracy_score(y_test, predicoes)
print("Acurácia do modelo:", acuracia)

# Visualizar a matriz de confusão
cm = confusion_matrix(y_test, predicoes)
print("Matriz de Confusão:")
print(cm)
```
